# Title
 The Hidden Risk of Unhygienic Health Facilities

# Introduction
Across the globe, millions rely on local clinics and hospitals for safe, life-saving care. Yet for many communities, these same “sanctuaries of health” can become sources of infection—simply because they lack the most fundamental provision: regular cleaning services.
This report shines a light on the scope and consequences of health facilities operating without routine environmental hygiene.

# Key Questions
How widespread is the absence of professional cleaning services in health facilities?

Which regions and facility types (clinics, health posts, hospitals) are most affected?

Is there a link between lack of cleaning and key health outcomes—e.g., infection rates or patient trust?

Have things improved or worsened over the past decade?

# Visualizations in This Report
 Global Map: % of facilities without cleaning services, by country.

 Bar Chart: Top 10 countries with the highest share of unclean facilities.

 Scatterplot: Relationship between cleaning-service coverage and reported facility-acquired infection rates.

 Time Series: Trends in cleaning-service availability for primary health clinics over the last 10 years.


Key Takeaways & Call to Action

Even small increases in routine cleaning can yield large health gains.

Urge stakeholders to integrate cleaning­­-service metrics into health-system funding, training, and reporting.

Point to success stories from countries that have scaled up environmental hygiene in rural clinics.

Conclusion
Unclean health facilities aren’t merely an inconvenience—they’re a direct threat to patient safety, staff well-being, and community trust.
By bringing cleaning services out of the shadows and into every clinic, we take a critical step toward safer, more resilient health systems for all.

```python
import pandas as pd

# Load the datasets
indicator_df = pd.read_csv('unicef_indicator_1.csv')
metadata_df = pd.read_csv('unicef_metadata.csv')

# Display to confirm
indicator_df.head(), metadata_df.head()
```

```python
# Filter and drop NAs
scatter_df = metadata_df.dropna(subset=[
    'GDP per capita (constant 2015 US$)',
    'Life expectancy at birth, total (years)'
])

# Use most recent year
scatter_latest = scatter_df.sort_values('year').groupby('country').last().reset_index()

# Plot
(ggplot(scatter_latest, aes(x='GDP per capita (constant 2015 US$)', # Removed backticks from x and y
                            y='Life expectancy at birth, total (years)')) + # Removed backticks from x and y
 geom_point(alpha=0.6) +
 geom_smooth(method='lm', color='red') +
 labs(title='GDP per Capita vs Life Expectancy',
      x='GDP per Capita (2015 USD)', y='Life Expectancy') +
 theme_minimal())
```

```python
from plotnine import *

# Filter the data
cleaning_df = indicator_df[
    (indicator_df['indicator'] == 'Proportion of health care facilities with no cleaning service') &
    (indicator_df['obs_value'].notna())
]

# Get latest available value per country
latest_cleaning = cleaning_df.sort_values('time_period') \
    .groupby('country').last().reset_index()

# Plot the bar chart
(ggplot(latest_cleaning, aes(x='reorder(country, obs_value)', y='obs_value')) +
 geom_col(fill='skyblue') +
 coord_flip() +
 labs(title='Health Facilities Without Cleaning Service (Latest Year)',
      x='Country', y='% without Cleaning Service') +
 theme_minimal())
```

```python
bhutan_cleaning = cleaning_df[cleaning_df['country'] == 'Bhutan']

# Plot time series
(ggplot(bhutan_cleaning, aes(x='time_period', y='obs_value')) +
 geom_line(color='mediumseagreen') +
 geom_point() +
 labs(title='Health Facilities Without Cleaning Service in Bhutan',
      x='Year', y='% without Cleaning Service') +
 theme_minimal())
```

```python
# Filter education data for the most recent year per country for total population
ed_latest = indicator_df[indicator_df['sex'] == 'Total'].copy()
ed_latest = ed_latest.sort_values('time_period', ascending=False)
ed_latest = ed_latest.drop_duplicates(subset=['country'], keep='first')

# Get the most recent metadata for each country
meta_latest = metadata_df.copy()
meta_latest = meta_latest.sort_values('year', ascending=False)
meta_latest = meta_latest.drop_duplicates(subset=['country'], keep='first')

# Merge datasets
merged_data = pd.merge(ed_latest, meta_latest, on='country', how='inner')

# Create a dataset for gender comparison
gender_data = indicator_df[indicator_df['sex'] != 'Total'].copy()
gender_data = gender_data.sort_values('time_period', ascending=False)
gender_data = gender_data.drop_duplicates(subset=['country', 'sex'], keep='first')

# Get time series data for countries with multiple years
time_data = indicator_df[indicator_df['sex'] == 'Total'].copy()
time_data = time_data.sort_values(['country', 'time_period'])

# Display the merged dataset
merged_data.head()

# Create a map-like visualization using a bubble chart approach
# First ensure we have the data we need
world_data = merged_data.copy()

# Create a mapping of regions to approximate x,y coordinates for visualization
# This creates an approximate geographic distribution
region_mapping = {
    'Africa': {'x_min': 0, 'x_max': 33, 'y_min': 0, 'y_max': 40},
    'Asia': {'x_min': 34, 'x_max': 70, 'y_min': 10, 'y_max': 50},
    'Europe': {'x_min': 34, 'x_max': 60, 'y_min': 51, 'y_max': 70},
    'North America': {'x_min': 71, 'x_max': 90, 'y_min': 40, 'y_max': 70},
    'South America': {'x_min': 71, 'x_max': 90, 'y_min': 0, 'y_max': 39},
    'Oceania': {'x_min': 91, 'x_max': 100, 'y_min': 0, 'y_max': 30}
}

# Function to assign approximate x,y coordinates based on country name
def assign_coordinates(country):
    import random

    # Simple region assignment based on country name
    # This can be improved with a proper lookup table
    if country in ['South Africa', 'Kenya', 'Nigeria', 'Ethiopia', 'Egypt']:
        region = 'Africa'
    elif country in ['China', 'India', 'Japan', 'Thailand', 'Afghanistan']:
        region = 'Asia'
    elif country in ['Germany', 'France', 'United Kingdom', 'Italy', 'Spain']:
        region = 'Europe'
    elif country in ['United States', 'Canada', 'Mexico']:
        region = 'North America'
    elif country in ['Brazil', 'Argentina', 'Chile', 'Peru']:
        region = 'South America'
    elif country in ['Australia', 'New Zealand']:
        region = 'Oceania'
    else:
        # Default assignment based on first letter of country name
        first_letter = country[0].upper()
        if first_letter <= 'F':
            region = 'Africa'
        elif first_letter <= 'L':
            region = 'Asia'
        elif first_letter <= 'Q':
            region = 'Europe'
        elif first_letter <= 'T':
            region = 'North America'
        elif first_letter <= 'Z':
            region = 'South America'
        else:
            region = 'Oceania'

    # Get region boundaries
    bounds = region_mapping[region]

    # Assign random coordinates within the region's boundaries
    x_coord = random.uniform(bounds['x_min'], bounds['x_max'])
    y_coord = random.uniform(bounds['y_min'], bounds['y_max'])

    return pd.Series([x_coord, y_coord, region])

# Apply the function to create x,y coordinates
world_data[['x_coord', 'y_coord', 'region']] = world_data.apply(
    lambda row: assign_coordinates(row['country']), axis=1
)

# Create the map visualization
(ggplot(world_data, aes(x='x_coord', y='y_coord', size='obs_value',
                       color='obs_value', label='country')) +
 geom_point(alpha=0.7) +
 # Add text labels for countries with high out-of-school rates
 geom_text(data=world_data[world_data['obs_value'] > 40],
           mapping=aes(label='country'),
           size=8,
           nudge_y=2,
           color='black') +
 scale_size(range=[2, 15]) +
 scale_color_gradient(low="#c6dbef", high="#084594") +
 labs(title="Geographic Distribution of Health Facilities without Cleaning Services",
      subtitle="Bubble size and color indicate Health Facilities without Cleaning Service percentage",
      x="",
      y="",
      size="rate (%)",
      color="rate (%)") +
 theme_minimal() +
 theme(figure_size=(12, 8),
       plot_title=element_text(size=14, face="bold"),
       axis_text_x=element_blank(),
       axis_text_y=element_blank(),
       axis_ticks=element_blank(),
       panel_grid_major=element_blank(),
       panel_grid_minor=element_blank()) +
 # Add region label annotations
 annotate("text", x=16.5, y=45, label="AFRICA", color="gray", size=12, alpha=0.3) +
 annotate("text", x=52, y=55, label="ASIA", color="gray", size=12, alpha=0.3) +
 annotate("text", x=47, y=75, label="EUROPE", color="gray", size=12, alpha=0.3) +
 annotate("text", x=80, y=75, label="NORTH AMERICA", color="gray", size=12, alpha=0.3) +
 annotate("text", x=80, y=35, label="SOUTH AMERICA", color="gray", size=12, alpha=0.3) +
 annotate("text", x=95, y=20, label="OCEANIA", color="gray", size=12, alpha=0.3))
```
